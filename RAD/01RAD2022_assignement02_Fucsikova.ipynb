* Otázka 2

  Sestavte jednoduchý regresní model a na jeho základech zjistěte zdali proměnná `indus`  ovlivňuje median ceny nemovitostí určených k bydlení `medv`. Pokud ano, o kolik je průměr mediánů cen nemovitostí nižší/vyšší při vzrůstu zastoupení nemaloobchodního podnikání o 5 jednotek? 

```r
model <- lm(medv ~ indus, Boston_filtered)
summary(model)

medv_change <- model$coefficients[2] * 5
medv_change
```

Ano, proměnná indus ovlivňuje medián ceny nemovitostí určených k bydlení, jelikož indus je v modelu statisticky významná, což ukazuje p-hodnota, která je menší než 0.05. Při vzrůstu zastoupení nemaloobchodního podnikání o 5 jednotek je průměr mediánů cen nemovitostí nižší o přibližně 2 680$.



* Otázka 3

  Vyzkoušejte model s mocninou a logaritmickou transformací odezvy. Pro výběr mocniné transformace vykreslete optimální  log-věrohodnostní profil u Box-Coxovy transformace a porovnejte navrženou transformaci s provedenou logaritmickou.

```r
model_log = lm(log(medv) ~ indus, Boston_filtered)
summary(model_log)
```

```r
bc <- boxcox(model, lambda = seq(-1,1 , 0.01))
lambda <- bc$x[which.max(bc$y)]
lambda

Boston_power <- Boston_filtered
Boston_power$medv_power <- (Boston_power$medv^lambda-1)/lambda

model_power <- lm(medv_power ~ indus, Boston_power)
summary(model_power)
```

```r
ggplot(Boston_filtered, aes(indus, log(medv)), ggtitle("Logaritmic transformation"))+geom_point() + stat_smooth(method = "lm", se = F, color = "blue", formula = y ~ x)
ggplot(Boston_power, aes(indus, medv_power), ggtitle("Boxcox transformation"))+geom_point() + stat_smooth(method = "lm", se = F, color = "blue", formula = y ~ x)
```

Z vykreslení je patrné, že jsou modely téměř shodné. Příčinou toho je, že se hodnota $\lambda$ příliš neliší od 0, která vede na logaritmickou transformaci.
