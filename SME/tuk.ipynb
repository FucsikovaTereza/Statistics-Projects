```python
#sestavení matice vysvětlujících proměnných a vektoru proměnné vysvětlované
A = np.transpose(df1.to_numpy())
X = sm.add_constant(np.transpose(np.delete(A,[0,1,2],0)))
y1 = A[1]
y2 = A[2]
```

```python
#vypsání výsledků pro linearní regresní model Brozka
model1 = sm.OLS(y1, X).fit()
print(model1.summary())
```

```python
#vypsání výsledků pro linearní regresní model Brozka po odstranění nevýznamných veličin
X1 = sm.add_constant(np.transpose(np.delete(A,[0,1,2,4,9,11,12,14,15,16,17],0)))
model1_2 = sm.OLS(y1, X1).fit()
print(model1_2.summary())
```

```python
#vypsání výsledků pro linearní regresní model Siri
model2 = sm.OLS(y2, X).fit()
print(model2.summary())
```

```python
#vypsání výsledků pro linearní regresní model Siri po odstranění nevýznamných veličin
X2 = sm.add_constant(np.transpose(np.delete(A,[0,1,2,9,10,11,12,15,14,16,17,18],0)))
model2_2 = sm.OLS(y2, X2).fit()
print(model2_2.summary())
```

```python
#výpočet reziduí
y1_pred = model1_2.predict(X1)
y2_pred = model2_2.predict(X2)

rez1 = y1 - y1_pred
rez2 = y2 - y2_pred
```

```python
#vykreslení reziduí v závislosti na predikci podle Brozka a Siri
fig, ax = plt.subplots(1, 2)

ax[0].plot(y1_pred, rez1, 'go')
ax[0].set_xlabel('predikce Y1')
ax[0].set_title('Brozek', fontsize = 14)

ax[1].plot(y2_pred, rez2, 'mo')
ax[1].set_xlabel('predikce Y2')
ax[1].set_title('Siri', fontsize = 14)

for i in range(2):
    ax[i].set_ylabel('rezidua')

fig.set_size_inches(10,4)
plt.savefig("rezidua_predikce.png")
```

```python
#vykreslení reziduí v závislosti na vysvětlujících proměnných podle Brozka
fig, ax = plt.subplots(3, 3)

ax[0,0].plot(A[3], rez1, 'go')
ax[0,0].set_xlabel('Hustota', fontsize = 17)

ax[0,1].plot(A[5], rez1, 'go')
ax[0,1].set_xlabel('Váha', fontsize = 17)

ax[0,2].plot(A[6], rez1, 'go')
ax[0,2].set_xlabel('Výška', fontsize = 17)

ax[1,0].plot(A[7], rez1, 'go')
ax[1,0].set_xlabel('Index tuku', fontsize = 17)

ax[1,1].plot(A[8], rez1, 'go')
ax[1,1].set_xlabel('Hmotnost bez tuku', fontsize = 17)

ax[1,2].plot(A[10], rez1, 'go')
ax[1,2].set_xlabel('Obvod hrudníku', fontsize = 17)

ax[2,0].plot(A[13], rez1, 'go')
ax[2,0].set_xlabel('Obvod stehna', fontsize = 17)
ax[2,0].set_ylabel('rezidua', fontsize = 17)

ax[2,1].plot(A[18], rez1, 'go')
ax[2,1].set_xlabel('Obvod zápěstí', fontsize = 17)
ax[2,1].set_ylabel('rezidua', fontsize = 17)

fig.delaxes(ax[2,2])         

for i in range(2):
    for j in range(3):
        ax[i,j].set_ylabel('rezidua', fontsize = 17)

fig.set_size_inches(19,16)
fig.suptitle('Brozek', fontsize = 25)
plt.savefig("rezidua_vysvetlujici_Brozek.png")
```

```python
#vykreslení reziduí v závislosti na vysvětlujících proměnných podle Siri
fig, ax = plt.subplots(3, 3)

ax[0,0].plot(A[3], rez2, 'mo')
ax[0,0].set_xlabel('Hustota', fontsize = 17)

ax[0,1].plot(A[4], rez2, 'mo')
ax[0,1].set_xlabel('Věk', fontsize = 17)

ax[0,2].plot(A[5], rez2, 'mo')
ax[0,2].set_xlabel('Váha', fontsize = 17)

ax[1,0].plot(A[6], rez2, 'mo')
ax[1,0].set_xlabel('Výška', fontsize = 17)

ax[1,1].plot(A[7], rez2, 'mo')
ax[1,1].set_xlabel('Index tuku', fontsize = 17)

ax[1,2].plot(A[8], rez2, 'mo')
ax[1,2].set_xlabel('Hmotnost bez tuku', fontsize = 17)

ax[2,0].plot(A[13], rez2, 'mo')
ax[2,0].set_xlabel('Obvod stehna', fontsize = 17)
ax[2,0].set_ylabel('rezidua', fontsize = 17)

fig.delaxes(ax[2,1]) 
fig.delaxes(ax[2,2])

for i in range(2):
    for j in range(3):
        ax[i,j].set_ylabel('rezidua', fontsize = 17)

fig.set_size_inches(19,16)
fig.suptitle('Siri', fontsize = 25)
plt.savefig("rezidua_vysvetlujici_Siri.png")
```
