## Datset

* **(A) Start temperature** (numeric variable): *low* 10 vs. *high* 50 \[°C]
* **(B) Salt concentration** (numeric variable): *low* 0 vs. *high* 1 \[%]
* **(C) Lid** (factor variable): *low* off vs. *high* on
* **(D) Water volume** (numeric variable): *low* 30 vs. *high* 70 \[% of pot volume]
* **(E) Heat degree** (numeric variable): *low* 5 *high* 9  
* **(F) Hotplate size** (factor variable): *low* 0 = small stove vs. *high* 1 = big stove
---
* **Response**: Time to reach a water temperature of 90°C (boiling temperature of 100°C was impossible to reach when the lid was off)
---

For our experiment, we're employing a $2^{6-2}$ factorial design with center points. This design incorporates four numeric and two factor variables. Accordingly, we require four center point measurements, leading to a total of 20 measurements.

```r
url = "https://gist.githubusercontent.com/FucsikovaTereza/82b773c5d91637273bd97d4f74c4fff6/raw/5bd731b2c44f9c85b9b80d060794bc1caef3b489/gistfile1.txt"
df_all = read.table(url, header=TRUE, sep="")
df_all <- df_all %>% transmute( start_temperature= as.factor(start_temperature),
                          salt_concentration = as.factor(salt_concentration),
                          lid = as.factor(lid),
                          water_volume = as.factor(water_volume),
                          heat_degree = as.factor(heat_degree),
                          hotplate_size = as.factor(hotplate_size),
                          boiling_time = as.numeric(boiling_time))

df_all
```

```r
# remove center points
df <- df_all[1:(nrow(df_all) - 4), ]
```

```r
k           <- 6
n_oper      <- 1

design      <- FrF2(2^(k-2), k, replications = n_oper, randomize = F, seed = c(48), generators = c("ABC","BCD"), factor.names = LETTERS[1:k])
time        <- df["boiling_time"]
design      <- add.response(design, time)

response.names(design)
design
```

```r
summary(design)
```

```r
par(mfrow=c(2,3))
boxplot(design$boiling_time ~ design$A, main = "Start temperature")
boxplot(design$boiling_time ~ design$B, main = "Salt concentration")
boxplot(design$boiling_time ~ design$C, main = "Lid")
boxplot(design$boiling_time ~ design$D, main = "Water volume")
boxplot(design$boiling_time ~ design$E, main = "Heat degree")
boxplot(design$boiling_time ~ design$F, main = "Hotplate size")
```

Box plots are used to visualize the distribution of the boiling time response variable across different levels of factors. The heat degree shows the most noticeable difference between the two levels. The high level has a much lower median boiling time and a more compact range, indicating that the boiling time is more consistent when the heat degree is high. Also, there are several points that are far above the upper whisker at the low level, suggesting potential outliers.

## Efects

```r
lm_for_coeffs <- lm(boiling_time ~ A*B*C*D*E*F, data=design)
effects <-as.data.frame(2*lm_for_coeffs$coefficients)
names(effects)[1] <- 'Effects'
effects
```


The table provided shows the estimated effects of the factors (and their interactions) on the boiling time from a linear model. Intercept represents the baseline level of the response variable, which is the average boiling time when all factors are at their reference levels (often the lowest level or the control level). If the value is negative, it suggests that increasing the factor from its low level to high level is associated with a decrease in the response variable (boiling time). We can see that e.g. B1:C1 (interaction between salt concentration and lid) is positive and relatively larger than other two-way interactions (0.88875).

```r
fit <- lm(boiling_time ~ A * B * C * D * E * F, data = design)
qqline(DanielPlot(fit, alpha = 0.05)$x)
```

In a normal probability plot, the residuals from the model (differences between observed and predicted boiling times) are plotted against a theoretical normal distribution. If the residuals are normally distributed, the points should roughly follow the straight line on the plot. A deviation from the line suggests that those factors or their interactions might have a significant effect on the response variable, which is boiling time. Any points that fall far from the straight line could be outliers or could indicate a non-linear relationship between those factors and the response. There is 90% confidence level used to draw the straight line in the plot. We should investigate A (start temperature), D(water volume), and E(heat degree) since they seem to have larger effects on boiling time.

```r
MEPlot(design)
```

The steepness of the slope indicates the strength of the main effect of that factor on the boiling time. The larger the absolute slope, the stronger the effect. Factors A and E show the most substantial changes in boiling time across their levels, suggesting they are important variables to consider in the process being studied. On the other hand for factor B there is almost no slope to the line, suggesting that the change in salt concentration has little to no effect on boiling time.

```r
IAPlot(design)
```

The lines for most interaction plots seem quite parallel or only slightly non-parallel, suggesting most factor interactions might not be significant.

## Models

```r
model_full <- lm(boiling_time ~ (.) + (.)*(.), data=design)
anova_result <- anova(model_full)
print(anova_result)
```

```r
# remove B:D
reduced_model <- lm(boiling_time ~ A + B + C + D + E + F + A:B + A:C + A:D + A:E + A:F + B:F, data=design)
summary(reduced_model)
```

```r
# remove B:D, A:C
reduced_model <- lm(boiling_time ~ A + B + C + D + E + F + A:B + A:D + A:E + A:F + B:F, data=design)
summary(reduced_model)
```

```r
# remove B:D, A:C, B
reduced_model <- lm(boiling_time ~ A + C + D + E + F + A:B + A:D + A:E + A:F + B:F, data=design)
summary(reduced_model)
```

```r
# remove B:D, A:C, B, F:B, A:B
reduced_model <- lm(boiling_time ~ A + C + D + E + F + A:D + A:E + A:F, data=design)
summary(reduced_model)
```

```r
par(mfrow = c(2, 2))
plot(reduced_model)
```


The diagnostic plots indicate that the linear regression model fits well overall, with residuals randomly dispersed around the horizontal line, suggesting appropriate linearity and homoscedasticity. However, there are outliers and potential points of leverage, particularly points labeled "2" and "12," which may impact the model's accuracy. While the residuals appear mostly normal, slight deviations exist.

```r
shapiro.test(residuals(reduced_model))
```

The value of W is 0.96758, which is the test statistic from the Shapiro-Wilk test. A value close to 1 suggests that the data is normally distributed. With such a high p-value, the test indicates that there is no reason to believe the residuals are not normally distributed (we don't reject the null hypothesis, that data is normally distributed).

```r
new_rows <- data.frame(matrix(c(0,0,-1,0,0,-1,4.24,0,0,-1,0,0,1,3.29,0,0,1,0,0,-1,4.18,0,0,1,0,0,1,3.21), nrow = 4, byrow = TRUE))
colnames(new_rows) <- colnames(design)

design$A <- factor(design$A, levels = c(-1, 0, 1))
design$B <- factor(design$B, levels = c(-1, 0, 1))
design$D <- factor(design$D, levels = c(-1, 0, 1))
design$E <- factor(design$E, levels = c(-1, 0, 1))

design_center <- rbind(design, new_rows)

design_center
```

```r
# remove B, A:B, B:D, B:F, C:F
model_center_2 <- lm(boiling_time ~ A + C + D + E + F + A:C + A:D + A:E + A:F, data = design_center)
summary(model_center_2)
```

```r
# remove A:C
model_center_3 <- lm(boiling_time ~ A + C + D + E + F + A:D + A:E + A:F, data = design_center)
summary(model_center_3)
```

```r
# remove F, A:F
model_center_4 <- lm(boiling_time ~ A + C + D + E + A:D + A:E, data = design_center)
summary(model_center_4)
```

```r
model_center_final <- model_center_3
```

```r
par(mfrow = c(2, 2))
plot(model_center_final)
```

The "Residuals vs Fitted" plot shows no clear patterns, indicating a good fit, but with a few outliers labeled "2" and "12" that could be influential. The "Q-Q Plot" suggests that residuals are mostly normally distributed, with some deviation at the extremes. The "Scale-Location" plot shows a relatively even spread of residuals across fitted values, implying consistent variance. The "Residuals vs Leverage" plot indicates that most data points do not have high leverage, but the point labeled "120" could be influential, warranting further investigation.

```r
shapiro.test(residuals(model_center_final))
```

The Shapiro-Wilk test shows a W statistic of 0.95214 and a p-value of 0.4007. This indicates that the null hypothesis of normality cannot be rejected at common significance levels (e.g., 0.05), suggesting that the residuals of the model are normally distributed. The model seems to meet the normality assumption for the residuals, which is a key consideration in many statistical modeling techniques.

```r
A_means <- aggregate(boiling_time ~ A, data = design_center, median)
A_line <- A_means[A_means$A %in% c(-1, 1), ]
D_means <- aggregate(boiling_time ~ D, data = design_center, median)
D_line <- D_means[D_means$D %in% c(-1, 1), ]
E_means <- aggregate(boiling_time ~ E, data = design_center, median)
E_line <- E_means[E_means$E %in% c(-1, 1), ]

plot_A <- ggplot(design_center, aes(x = A, y = boiling_time)) +
  geom_point() +
  geom_point(data = A_means, aes(x = A, y = boiling_time), color = "blue", size = 3) +
  geom_line(data = A_line, aes(group = 1), color = "red") +
  labs(title = "Boiling time vs. Start temperature", x = "Start temperature", y = "Boiling Time") +
  theme_minimal()

plot_D <- ggplot(design_center, aes(x = D, y = boiling_time)) +
  geom_point() +
  geom_point(data = D_means, aes(x = D, y = boiling_time), color = "blue", size = 3) +
  geom_line(data = D_line, aes(group = 1), color = "red") +
  labs(title = "Boiling Time vs. Water volume", x = "Water volume", y = "Boiling Time") +
  theme_minimal()

plot_E <- ggplot(design_center, aes(x = E, y = boiling_time)) +
  geom_point() +
  geom_point(data = E_means, aes(x = E, y = boiling_time), color = "blue", size = 3) +
  geom_line(data = E_line, aes(group = 1), color = "red") +
  labs(title = "Boiling Time vs. Heat degree", x = "Heat degree", y = "Boiling Time") +
  theme_minimal()

grid.arrange(plot_A, plot_D, plot_E, ncol = 3)
```

Overall, the model seems to capture linear trends between boiling time and both start temperature and heat degree, with the center points supporting these relationships. Water volume doesn't show a clear linear trend with boiling time based on the center points.

## Contour plots

```r
design_center$A <- ifelse(as.numeric(design_center$A) == 1, -1, 1)
design_center$B <- ifelse(as.numeric(design_center$B) == 1, -1, 1)
design_center$D <- ifelse(as.numeric(design_center$D) == 1, -1, 1)
design_center$E <- ifelse(as.numeric(design_center$E) == 1, -1, 1)
model_contour <- lm(boiling_time ~ A + C + D + E + F + A:D + A:E + A:F, data = design_center)
anova(model_contour)
```

```r
summary(model_contour)
```

```r
A_values <- seq(from = min(design_center$A), to = max(design_center$A), length.out = 100)
C_values <- factor(1)
D_values <- seq(from = min(design_center$D), to = max(design_center$D), length.out = 100)
E_values <- seq(from = min(design_center$E), to = max(design_center$E), length.out = 100)
F_values <- factor(1)

grid <- expand.grid(A = A_values, C = C_values, D = D_values, E = E_values, F = F_values)
grid$boiling_time <- predict(model_contour, newdata = grid)

# A:F was most significant, it is same as D:E
contour_plot <- ggplot(grid, aes(x = D, y = E, z = boiling_time)) +
  geom_contour_filled(aes(fill = ..level..)) +
  geom_contour(colour = "white") +
  labs(title = "Contour Plot of Boiling Time by Water volume and Heat degree (bigger hotplate)", x = "Water volume", y = "Heat degree", fill = "Boiling Time") +
  theme_minimal()

C_values <- factor(1)
F_values <- factor(-1)
grid2 <- expand.grid(A = A_values, C = C_values, D = D_values, E = E_values, F = F_values)
grid2$boiling_time <- predict(model_contour, newdata = grid2)

contour_plot2 <- ggplot(grid2, aes(x = D, y = E, z = boiling_time)) +
  geom_contour_filled(aes(fill = ..level..)) +
  geom_contour(colour = "white") +
  labs(title = "Contour Plot of Boiling Time by Water volume and Heat Degree (smaller hotplate)", x = "Water volume", y = "Heat degree", fill = "Boiling Time") +
  theme_minimal()


options(repr.plot.width = 16, repr.plot.height = 6)
grid.arrange(contour_plot, contour_plot2, ncol = 2)
```


The two contour plots illustrate the relationship between water volume, heat degree, and boiling time on different hotplate sizes. Dark purple areas represent the shortest boiling times, with the left plot erroneously indicating negative boiling times for a larger hotplate and the right plot correctly showing boiling times between 1 and 2 minutes for a smaller hotplate. The color gradients suggest that boiling time increases with greater water volume and lower heat degrees.

```r
optimal <- grid[which.min(grid$boiling_time), ]
print("Optimal conditions for D and E with F on high level (bigger hotplate):")
print(optimal)

optimal2 <- grid2[which.min(grid2$boiling_time), ]
print("Optimal conditions for D and E with F on low level (smaller hotplate):")
print(optimal2)
```

## Conclusion

Our findings indicate that the most critical variables affecting the boiling time of water are the intensity of heat applied, the volume of water, the initial water temperature, and the size of the heating surface. The presence of a lid on the pot also notably influenced boiling times. Salt concentration, in contrast, did not have a significant effect, which may be attributed to its relatively small range or its lesser impact compared to the other variables.
